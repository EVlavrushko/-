Написать приложение или функцию, которые собирают основные новости с сайта на выбор lenta.ru, yandex-новости. Для парсинга использовать XPath.
Структура данных в виде словаря должна содержать:
- *название источника;
- наименование новости;
- ссылку на новость;
- дата публикации.

```python
!pip install lxml
!pip install requests
from pprint import pprint
from lxml import html
import requests
import pandas as pd
```

```python
url = 'https://yandex.ru/news/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36', #Из-за неверного определения User-agent код ранее не работал. Простой способ - в инструментах разработчика выбрать сеть(network) и в любом запросе внизу указан юзерагент
}

params = {
    'sso_failed': '',
}
```

```python
response = requests.get(url, headers=headers, params=params)
content_dom = html.fromstring(response.text)
print(response.url)
```

```python
news_container = content_dom.xpath("//section//div[contains(@class, 'mg-grid__item')]")
yandex_news = []
for new_container in news_container:
  new_source = new_container.xpath(".//div[@class='mg-card-footer__left']//a/text()")[0]
  new_text = new_container.xpath(".//h2/a/text()")[0].replace('\xa0', ' ')
  new_link = new_container.xpath(".//h2/a/@href")[0]
  new_date = new_container.xpath(".//span[@class='mg-card-source__time']/text()")[0]
  news_dict = {
  'new_source': new_source,
  'new_text': new_text,
  'new_link': new_link,
  'new_date': new_date,
  }
  yandex_news.append(news_dict)
```

```python
pprint(yandex_news)
```

```python
#Заимствованный код
pd.DataFrame(yandex_news)
```
